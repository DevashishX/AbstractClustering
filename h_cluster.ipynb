```python
import glob, json

j = json.load(open('umap.json'))
```

```python
from hdbscan import HDBSCAN
```

```python
kwargs = {'vecs': j}
```

```python
config = {
  'min_cluster_size': 20,
  'cluster_selection_epsilon': 0.01,
  'min_samples': 1,
}
print(config)
z = HDBSCAN(**config).fit(kwargs['vecs'])
```

```python
len(kwargs['vecs'])**(0.5)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

ids = set(z.labels_)
print(' * found', len(ids), 'ids')
colors = [cm.rainbow(i/max(ids)) for i in z.labels_]
vecs = np.array(kwargs['vecs'])
x, y = vecs.T
plt.scatter(x, y, c=colors, s=0.01)
```
